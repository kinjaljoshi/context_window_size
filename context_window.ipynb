```python
models = {
    "Mistral": "mistralai/Mistral-7B-Instruct-v0.3",
    "LLaMA": "meta-llama/Llama-3.2-1B",
    "Qwen": "Qwen/Qwen2.5-VL-7B-Instruct"
}
```

```python
from transformers import AutoTokenizer

tokenizers = {name: AutoTokenizer.from_pretrained(model) for name, model in models.items()}
for name, tokenizer in tokenizers.items():
    print(f"{name} Context Window: {tokenizer.model_max_length} tokens")
```

Context Window: 1000000000000000019884624838656 tokens

It likely means one of the following issues:

* Integer Overflow or Placeholder Value
Some models return very large numbers as a default placeholder for model_max_length because they don’t have a strict limit.
This can happen when Google's Gemma tokenizer doesn't explicitly define the max context size in Hugging Face.
* Tokenizer Not Properly Configured
If a tokenizer doesn’t define model_max_length, it may default to int(1e30), which is too large.


```python
text = "Artificial intelligence is revolutionizing the world."

for name, tokenizer in tokenizers.items():
    tokens = tokenizer(text, return_tensors="pt")
    print(f"\n{name} Tokenized Output: {tokens['input_ids'].shape[1]} tokens")
```

```python
for name, tokenizer in tokenizers.items():
    bos_token = tokenizer.bos_token
    eos_token = tokenizer.eos_token
    print(f"\n{name} BOS Token: {bos_token}, EOS Token: {eos_token}")
```

```python
for name, tokenizer in tokenizers.items():
    tokenized_text = tokenizer.tokenize(text)
    print(f"\n{name} Tokenized Words: {tokenized_text}")
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("google/gemma-3-4b-it")
name = "Gemma-3"
print(f"{name} Context Window: {tokenizer.model_max_length} tokens")

text = "Artificial intelligence is revolutionizing the world."

tokens = tokenizer(text, return_tensors="pt")
print(f"\n{name} Tokenized Output: {tokens['input_ids'].shape[1]} tokens")
bos_token = tokenizer.bos_token
eos_token = tokenizer.eos_token
print(f"\n{name} BOS Token: {bos_token}, EOS Token: {eos_token}")
tokenized_text = tokenizer.tokenize(text)
print(f"\n{name} Tokenized Words: {tokenized_text}")
```
